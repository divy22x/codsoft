```python
unique_areas = df['Area'].unique()

max_occurrence = 0
area_with_max_occurrence = None
for area in unique_areas:
    area_data = df[df['Area'] == area]
    gender_counts = area_data['Gender'].value_counts()

    total_occurrence = gender_counts.sum()

    if total_occurrence > max_occurrence:
        max_occurrence = total_occurrence
        area_with_max_occurrence = area

    print(f'Gender Distribution in {area}:')
    print(gender_counts)

    sns.set(style="whitegrid")
    plt.figure(figsize=(6, 4))
    ax = sns.countplot(x='Gender', data=area_data)

    total = float(len(area_data))
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x() + p.get_width() / 2., height + 0.1,
                '{:.1%}'.format(height / total),
                ha="center")

    plt.title(f'Gender Distribution in {area}')
    plt.show()

print(f'The area with the highest occurrence is: {area_with_max_occurrence} ({max_occurrence} occurrences)')
```

```python
def analyze_area(df, target_area):

    area_data = df[df['Area'] == target_area]

    if area_data.empty:
        print(f"No data found for {target_area}.")
        return

    gender_counts = area_data['Gender'].value_counts()

    print(f'Gender Distribution in {target_area}:')
    print(gender_counts)

    sns.set(style="whitegrid")
    plt.figure(figsize=(6, 4))
    ax = sns.countplot(x='Gender', data=area_data)

    total = float(len(area_data))
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x() + p.get_width() / 2., height + 0.1,
                '{:.1%}'.format(height / total),
                ha="center")

    plt.title(f'Gender Distribution in {target_area}')
    plt.show()

target_area = input("Enter the area you want to analyze: ")
analyze_area(df, target_area)
```
